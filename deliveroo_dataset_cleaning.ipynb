```python
def clean_column_names(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.replace(r'[^\w\s]', '', regex=True)
        .str.replace(r'\s+', '_', regex=True)
        .str.lower()
    )

    df = df.rename(columns={
        'mealname': 'meal_name',
        'addressa': 'address',
        'lprice_in_euro': 'price_in_euro',
        'totalrevenue': 'total_revenue'
    })

    return df


deliveroo_df = clean_column_names(deliveroo_df)


print(deliveroo_df.columns)
```

```python
def order_time_function(row):
        if int(row) > 4 and int(row) < 12:
            return "morning"
        elif int(row) == 12:
            return "noon"
        elif int(row) > 12 and int(row) < 18:
            return "afternoon"
        elif int(row) >= 18 and int(row) < 22:
            return "evening"
        elif int(row) >= 22:
            return "night"
        elif int(row) <= 4:
            return "late_night"
        else:
             return row
    

deliveroo_df["day_period"] = deliveroo_df["order_time"].apply(order_time_function)

deliveroo_df["day_period"].value_counts()


```

```python
deliveroo_df['parsed_date'] = pd.to_datetime(deliveroo_df['order_dates'])


deliveroo_df['weekday'] = deliveroo_df['parsed_date'].dt.day_name()
deliveroo_df['date'] = deliveroo_df['parsed_date'].dt.strftime('%d-%m-%Y')


deliveroo_df.drop(columns=['order_dates', 'parsed_date'], inplace=True)
```

```python
columns_to_remove = [
    "complaints",
    "desired_temperature",
    "tax_sum",
    "refunded",
    "total_revenue",
    "delivery_temperature"
]

df_cleaned = deliveroo_df.drop(columns=columns_to_remove)
df_cleaned = df_cleaned.rename(columns={"price_in_euro": "price_in_order"})
df_cleaned.head(2)
```

```python
df_cleaned["order_id"] = ["order_" + str(i) for i in df_cleaned.index]
df_cleaned = df_cleaned[["order_id"] + [col for col in df_cleaned.columns if col != "order_id"]]


df_cleaned.head(2)
```

```python
from unidecode import unidecode

def clean_unicode(x):
    return unidecode(str(x)) if isinstance(x, str) else x

df_cleaned_ascii = df_cleaned.applymap(clean_unicode)
```

```python
df_cleaned_ascii.to_csv(
    "cleaned_ascii.csv",
    index=False,
    encoding="utf-8",
    sep=",",
    quotechar='"',
    quoting=1
)
```
